```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the csvl file
df_full_variables = pd.read_csv('Data+Design_Full variable.csv')
df_urban_growth = pd.read_csv('Urban_Growth.csv')
 
# Display the first few rows of each dataframe
df_full_variables.head(), df_urban_growth.head()
```

```python
# Display the info
df_full_variables.head()
```

```python
# Select populous countries for visualization
Populous_countries = ["Nigeria", "Ethiopia", "Egypt", "Tanzania", "South Africa", "Kenya", "Ghana"]

# To Filter data for selected countries in the "urban_growth" sheet
df_urban_choice = df_urban_growth[df_urban_growth["countrywb"].isin(Populous_countries)]

# To Melt the dataframe for better visualization
df_urban_choice_melted = df_urban_choice.melt(id_vars=["countrywb"], var_name="Year", value_name="Urban Growth Rate")

# Convert Year column to integer
df_urban_choice_melted["Year"] = df_urban_choice_melted["Year"].astype(int)

# Plot urban growth trends
plt.figure(figsize=(12, 6))
for country in Populous_countries:
    country_data = df_urban_choice_melted[df_urban_choice_melted["countrywb"] == country]
    plt.plot(country_data["Year"], country_data["Urban Growth Rate"], marker="o", label=country)

plt.xlabel("Year")
plt.ylabel("Urban Growth Rate (%)")
plt.title("Trend of Urban Growth Rate in African Countries (2000-2021)")
plt.legend(loc="upper right")
plt.grid(True)
plt.show()
```

## Story telling
The rate of urbanization can provide insights into the socio-economic development of the area (countries). The chart above illustrates the urban growth rates of some African countries—"Nigeria", "Ethiopia", "Egypt", "Tanzania", "South Africa", "Kenya", "Ghana"—between 2000 and 2021.

The key insights from this charts include:

Tanzania, Ethiopia, Egypt and Nigeria show high urban growth rates,with slight decline after 2015. Possible due to migration trends, government policies or natural disasters. In the case of Nigeria, this period was significantly marked by increased insurgency and instability in the country.South Africa show relatively lower and more stable urban growth, possibly due to their already high urbanization levels.

SouthAfrica showed a sharp dip around 2015 - 2017, which may indicate sharp policy chnage or mass migration. 

Overall, the chart suggests that Africa Urbanization growth has been slow in the last 10years, suggesting economic instability, mass migration and socio-political issues.
