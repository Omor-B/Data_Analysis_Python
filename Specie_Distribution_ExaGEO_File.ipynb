```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('ExaGEO Exercise_Penguins.csv')
data.head()
```

```python
data.info()
```

```python
#Droping of rows with missing values from all the columns 
data = data.dropna(subset=['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex'])
```

```python
data.info()
```

```python
# # Visualising Culmen Length Across Species using barplot visualization by Creating mean and variance for culmen length across species
plt.figure(figsize=(10, 6))
sns.barplot(x='species', y='bill_length_mm', hue='species', data=data, palette='Set2',legend=False)
plt.title('Mean and Variance of Culmen Length Across Species', fontsize=16)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Culmen Length (mm)', fontsize=12)
plt.tight_layout()
plt.show()

```

```python
# Create a correlation matrix to explore the relationship between culmen length and other numerical features
correlation_data = data[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]
correlation_matrix = correlation_data.corr()
```

```python
#Correlation Analysis by Creating a heatmap to show the correlation
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Between Numerical Features', fontsize=16)
plt.tight_layout()
# plt.show()
```

```python
# ploting of Pairplots for detailed analysis between culmen length and other numerical features
sns.pairplot(data[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'species']], hue='species')
plt.suptitle('Pairplot of Numerical Features Across Species', y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Plot histograms to visualize the distribution of numerical features
plt.figure(figsize=(12, 10))

# Bill Length Distribution
plt.subplot(2, 2, 1)
sns.histplot(data['bill_length_mm'], kde=True, color='skyblue', bins=20)
plt.title('Distribution of Bill Length')

# Flipper Length Distribution
plt.subplot(2, 2, 2)
sns.histplot(data['flipper_length_mm'], kde=True, color='salmon', bins=20)
plt.title('Distribution of Flipper Length')

# Body Mass Distribution
plt.subplot(2, 2, 3)
sns.histplot(data['body_mass_g'], kde=True, color='green', bins=20)
plt.title('Distribution of Body Mass')

plt.tight_layout()
plt.show()
```

```python
# Visualize the distribution of sex across species
plt.figure(figsize=(8, 6))
sns.countplot(x='species', hue='sex', data=data, palette='rainbow')
plt.title('Sex Distribution Across Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```
